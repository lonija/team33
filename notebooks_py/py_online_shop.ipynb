
Develop a simple prototype for an online shop
1. Define classes Client, Items, Transaction
2. Write the logic so that each client can purchase one of several items in a single transaction
3. Make a printout of data with nested loops
4. The program should print out:
- all clients
- all transactions of a client
- all items purchased in each transaction of the client

```python
class Client:
    number_of_clients = 0

    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.transactions = [] #list of transactions is created here
        Client.number_of_clients += 1

    def add_transaction(self, transaction):
        self.transactions.append(transaction)  #transaction is defined under Clients

def view_clients(clients):
    print("All clients:")
    for client in clients:
        print(f"   ID: {client.id}, client name: {client.name}")

class Item:
    number_of_items = 0

    def __init__(self, item_id, title, cost, stock):
        self.item_id = item_id
        self.title = title
        self.cost = cost
        self.stock = stock
        Item.number_of_items += 1

    def update_stock(self, quantity):
        self.stock += quantity

class Transaction:
    number_of_transactions = 0

    def __init__(self, client, item, quantity):
        self.client = client
        self.item = item
        self.quantity = quantity
        Transaction.number_of_transactions += 1
        self.process_transaction()

    def process_transaction(self):
        if self.item.stock >= self.quantity:
            self.item.update_stock(-self.quantity)
            self.client.add_transaction(self)
            print(f"{self.client.name} has purchased {self.quantity} pcs of {self.item.title}.")
        else:
            print(f"Transaction unseccessful: insufficient stock of {self.item.title}.")

def print_transactions(clients):
    for client in clients:
        print(f"Transactions of {client.name}:")
        for transaction in client.transactions:
            print(f"  Transaction: item: {transaction.item.title}, quantity: {transaction.quantity}.")

clients = []
clients.append(Client("98567e", "Anna"))
clients.append(Client("234y67", "Joan"))
clients.append(Client("7d3046", "Mimi"))

items = []
items.append(Item("0234", "Laptop", 1000, 1))
items.append(Item("1202", "Smartphone", 700, 5))
items.append(Item("3419", "Book", 10, 1562))
transactions = []
transactions.append(Transaction(clients[0], items[0], 1))
transactions.append(Transaction(clients[1], items[1], 3))
transactions.append(Transaction(clients[2], items[0], 1))
transactions.append(Transaction(clients[2], items[2], 12))

#print out of the number of clients and their transactions
print(f"Total number of clients: {Client.number_of_clients}")
view_clients(clients)
for client in clients:
    print(f"Client {client.name} has made {len(client.transactions)} transaction(s).")

print_transactions(clients)
```
