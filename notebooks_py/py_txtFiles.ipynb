```python
#natural population growth

with open("/content/births.txt", "r") as file:
  births = file.readlines()
with open("/content/deaths.txt", "r") as file:
  deaths = file.readlines()

total_sum_births = 0
total_sum_deaths = 0

for birthline in births:
    total_sum_births += int(birthline[4:])
for deathline in deaths:
    total_sum_deaths += int(birthline[4:])
natural_growth = total_sum_births - total_sum_deaths
print("Natural growth is: ", natural_growth)
```

```python
#natural population growth

with open("/content/births.txt", "r") as file:
  births = file.readlines()
with open("/content/deaths.txt", "r") as file:
  deaths = file.readlines()

for i in range(len(births)):
  birthline = births[i]
  deathline = deaths[i]

  birthnumber = int(birthline[4:])
  deathnumber = int(deathline[4:])

  growth = birthnumber - deathnumber
  month = birthline[:3]
  #print("Natural growth in ", month, "is", growth)

  if growth > 0:
    print(f"Natural growth in {month} is {growth}")
```

```python
import pandas as pd

births = pd.read_csv("/content/births.txt", header = None, delimiter = " ")

birth_column_names = ["month", "number of births"]
births.columns = birth_column_names
births
```

```python
import pandas as pd

deaths = pd.read_csv("/content/deaths.txt", header = None, delimiter = " ")

death_column_names = ["month", "number of deaths"]
deaths.columns = death_column_names
deaths
```

```python
#merging

births["Diff. between births and deaths"] = births["number of births"] - deaths["number of deaths"]
births
```

```python
file_name = "output.csv"
births.to_csv(file_name, index = False)
```
