**DATA TYPES**

```python
#the data type is set when you assign a value to a variable
x = 1j #complex
x = ["apple", "banana", "cherry"] #list
x = ("apple", "banana", "cherry") #tuple
x = range(6) #range
x = {"name" : "John", "age" : 36} #dict
x = {"apple", "banana", "cherry"} #set
x = frozenset({"apple", "banana", "cherry"}) #frozen set
x = b"Hello" #bytes
x = None #none type

#you can specify the data type
x = str("Hello World")
x = dict(name="John", age=36)

#Variables do not need to be declared with any particular type,
#and can even change type after they have been set.
x = 4       # x is of type int
x = "Sally" # x is now of type str
print(x)

#If you want to specify the data type of a variable, this can be done with casting.
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0

#You can get the data type of a variable with the type() function.
x = 5
y = "John"
print(type(x))
print(type(y))
```

**ARITHMETIC CALCULATIONS**

```python
#addition
a = 5
b = 7
c = -2
sum = a + b + c
print(sum)
type(sum)
print(f'SUM of numbers a, b, c: {sum}')

#substraction
difference = a - b
print(difference)
print(f'DIFFERENCE between numbers a, b: {difference}')

#multiplication
multipl = a * b * c
print(f'MULTIPLICATION of numbers a, b, c: {multipl}')

#division
divided = a / b
print(f'DIVISION of numbers a, b, rounded: {round(divided, 2)}')

#exponentiation - a^b
power = a ** b
print(f'Exponentiation of a and b:  {power}')
```

```python
#DENSITY OF AN AREA
area = input("Enter area in sq km: ")
area_int = int(area) #could have been done before
#area = int(input("Enter area in sq km: "))
population = input("Enter population: ")
population_int = int(population)

density = round(population_int / area_int)
print(f"Density is: {density} people per sq km")
```

```python
#CONVERTING POUNDS TO KILOGRAMS
#a pound is 0.45359237 kg
pounds = float(input("Enter the weight in pounds: "))
conversion_factor = 0.45359237 #constant variable
kilograms_rounded = round(pounds * conversion_factor, 2)
print(f"Weight in pounds is: {pounds} lb, weight in kilograms (rounded) is: {kilograms_rounded} kg.")
```
