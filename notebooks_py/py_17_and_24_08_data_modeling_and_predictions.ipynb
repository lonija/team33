```python
import pandas as pd
import seaborn as sns
```

```python
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
```

DATA INVESTIGATION

```python
#how many empty values are there?
titanic_data.isnull().sum()
```

```python
titanic_data.drop("deck", axis = 1, inplace = True) #dropping column with too many NULL
```

```python
titanic_data.dtypes #what data types are there?
```

DATA TRANSFORMATION

```python
titanic_data["class"] = titanic_data["class"].astype("str") #changing data type to string

titanic_data.fillna(0, inplace = True) #filling empty values with 0
```

```python
#dropping not needed columns

titanic_data.drop(["embark_town", "alive", "alone", "adult_male", "class"], axis = 1, inplace = True)
```

```python
titanic_data.head()
```

```python
titanic_data["sex"].unique() #"male", "female"
```

```python
#create dictionary
sex_mapping = {"male": 0, "female": 1}

#replace values
titanic_data["sex"] = titanic_data["sex"].map(sex_mapping)
```

```python
#embark town
titanic_data["embarked"].unique() #S, C, Q

embark_town_mapping = {"S": 0, "C": 1, "Q": 2, 0:0}

#replace
titanic_data["embarked"] = titanic_data["embarked"].map(embark_town_mapping)
```

```python
#embarked

titanic_data["embarked"].unique()

#mapping
embarked_mapping = {"S": 0, "C": 1, "Q": 2, 0:0}

#replace
titanic_data["embarked"] = titanic_data["embarked"].map(embarked_mapping)
```

```python
#who

titanic_data["who"].unique() #child, woman, man

who_mapping = {"child": 0, "woman": 1, "man": 2, 0:0}

#replace
titanic_data["who"] = titanic_data["who"].map(who_mapping)
```

```python
titanic_data.head()
```

```python
#data model
#splitting the data for training and testing

from sklearn.model_selection import train_test_split

X = titanic_data.drop("survived", axis = 1) #features
Y = titanic_data["survived"] #dependant variable
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7)
```

LOGISTICS REGRESSION

```python
#import the module
from sklearn.linear_model import LogisticRegression

#initialize the model
model = LogisticRegression()

#train the model
model.fit(X_train, Y_train)

#predict using the model
prediction_results = model.predict(X_test)

#print the prediction results
print(prediction_results)
```
